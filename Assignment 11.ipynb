```python
# wap ask the user take the radius find the area of circle
    # pi=method use 
    # r=100
   # area_of_circle=pi*r*r

import math
pi=math.pi

def area_of_circle():
    radius=eval(input('enter the radius:'))
    aoc=pi*radius*radius
    return aoc
    
area_of_circle()
```

```python
# wap ask the user take height and breadth find the area of traingle
#       area_of_traingle=0.5*b*h

def area_of_triangle():    
    height=eval(input('enter the height:'))
    breadth=eval(input('enter the breadth:'))
    area=0.5*breadth*height
    return area

print(f'The area of triangle is {area_of_triangle()}')   
```

```python
# wap ask the user take 3 numbers find the average

def Average():
    n1=eval(input('enter the first number:'))
    n2=eval(input('enter the second number:'))
    n3=eval(input('enter the third number:'))
    avg=(n1+n2+n3)/3
    return avg

print(f'The Average of three numbers is {Average()}')
```

```python
# wap ask the user take the bill amount and tip amount calculate total bill

def Total_bill():
    bill_amount=eval(input('enter the bill amount:'))
    tip=eval(input('enter the tip amount:'))
    total_bill=bill_amount+tip
    return total_bill

print(f'The Total bill would be {Total_bill()}rs')
```

```python
# wap ask the user take the salary, percetage of tax calculate tax amount

def Tax_cal():
    salary=eval(input('enter the salary:'))
    tax_percent=eval(input('enter the tax_percentage:'))
    tax_amount=salary*tax_percent/100
    return tax_amount

print(f'The Tax amount would be {Tax_cal()}rs')
```

```python
# wap ask the user take the value in dollar convert into rupees
#         1$=80rs

def currency_conversion():
    dollar=eval(input('enter the amount in dollars:'))
    conversion=dollar*80
    return conversion

print(f'After conversion the amount would be {currency_conversion()}rs')
```

```python
# wap ask the user take weight in kgs convert into pounds
#     1kg=2.2pounds

def weight_conversion():
    weight=eval(input('enter the weight in kilograms:'))
    pounds=weight*2.2
    return pounds

print(f'The weight in pounds is {weight_conversion():.2f}')
    
```
