```python
# Q1: Sum of Three Numbers
# Write a function `sum_three(a, b, c)` that returns the sum of the three numbers.
# Example:
# sum_three(3, 5, 7) ➞ 15

def sot(a,b,c):
    summ=a+b+c
    return summ

sot(3,5,7)

```

```python
# Q2: Area of a Circle
# Write a function `area_circle(radius)` that returns the area of a circle. Use π = 3.14159
# Example:
# area_circle(7) ➞ 153.9381

import math
pi=math.pi

def area_of_circle(radius):
    aoc=pi*radius*radius
    return round(aoc,2)

area_of_circle(7)
```

```python
# Q3: Check Even or Odd
# Write a function `check_even_odd(num)` that prints whether a number is Even or Odd.
# Example:
# check_even_odd(5) ➞ &quot;Odd&quot;
# check_even_odd(8) ➞ &quot;Even&quot;



def eod(num):
    if num%2==0:
        print('even')
    else:
        print('odd')
    return num

eod(8)
```

```python
# Q4: Calculate BMI
# Write a function `calculate_bmi(weight, height)` that returns BMI value using the formula:
# BMI = weight / (height ** 2)
# Example:
# calculate_bmi(70, 1.75) ➞ 22.86


def bmi_cal(weight, height):
    bmi=weight/(height**2)
    return round(bmi,2)

bmi_cal(56, 1.70)
```

```python
# Q5: Greet with Name and Age
# Write a function `greet(name, age)` that prints:
# Hello &lt;name&gt;, you are &lt;age&gt; years old!

# Example:
# greet('Omkar', 30) ➞ Hello Omkar, you are 30 years old!


def greet(name, age):
    print(f'hello {name}, you are {age} years old')

greet('vijay', 24)
    
```

```python
# Q6: Find Maximum of Two Numbers
# Write a function `find_max(a, b)` that returns the greater of the two numbers.
# Example:
# find_max(4, 9) ➞ 9

def find_max(a,b):
    if a>b:
        return a
    else:
        return b

find_max(10,20)
```

```python
# Q7: Power Function
# Write a function `power(base, exponent)` that returns base raised to the power exponent.
# Example:
# power(2, 3) ➞ 8

def power(base, exponent):
    power=base**exponent
    return power

power(2, 3)
```

```python
# Q8: Calculate Simple Interest
# Write a function `simple_interest(principal, rate, time)` that returns the simple interest.
# Formula: (P × R × T) / 100
# Example:
# simple_interest(1000, 5, 2) ➞ 100.0


def si(principle, rate, time):
    interest=principle*rate*time/100
    return interest

si(1000, 5, 2)
```
