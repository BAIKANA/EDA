```python
# 4.Print the of a number (e.g., 374 → 3+7+4) using only loops.

number=eval(input('enter the number:'))
sum_digit=0

while number>0:
    last_digit=number%10
    sum_digit=sum_digit+last_digit
    number=number//10

print('sum of digits:', sum_digit)
```
